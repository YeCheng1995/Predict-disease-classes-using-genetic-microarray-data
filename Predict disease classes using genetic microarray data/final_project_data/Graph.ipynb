```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator

datas = pd.read_csv('./summary.csv')
```

```python
x = ['2','4','6','8','10','12','15','20','25','30']
y = datas.iloc[0]
y = y.tolist()
del(y[0])
y_float = []
for num in y:
    y_float.append(float(num))

plt.figure(figsize=(16, 8), dpi=80)
plt.title('Relationship between error rate and number of genes')
plt.xlabel('Number of genes')
plt.ylabel('Ave err rate')
plt.plot(x,y_float)
plt.show()
```
